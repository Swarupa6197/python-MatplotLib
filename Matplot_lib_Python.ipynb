```python
import matplotlib.pyplot as plt
```

```python
#example data
x=[10,20,30,40]
y=[5,6,7,8]

#plot a single line
plt.plot(x,y)

#give labels to axis
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')

#give title
plt.title('Basic Line Plot')

#display graph
plt.show()
```

```python
#Linear graph
months=['Jan','Feb','Mar','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
sales=[350,370,300,280,310,530,500,290,460,600,230,700]
plt.plot(months,sales,marker='*',color='blue',linestyle='--')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales in 2025')
plt.show()
```

```python
# bar graph
months=['Jan','Feb','Mar','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
sales=[350,370,300,280,310,530,500,290,460,600,230,700]
plt.bar(months,sales,color='pink',width=0.5)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales in 2025')
plt.show()
```

```python
# horizontal bar graph
months=['Jan','Feb','Mar','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
sales=[350,370,300,280,310,530,500,290,460,600,230,700]
plt.barh(sales,months,color='skyblue',height=0.5)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales in 2025')
plt.show()
```

```python
#histogram
plt.hist(months,bins=10,edgecolor='black',color='blue')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Monthly Sales in 2025')
plt.show()
```

```python
#histogram plotting using random data
import numpy as np
np.random.randint(40,100,size=50)
plt.hist(months,bins=10,edgecolor='black',color='pink')
plt.xlabel('Score_Range')
plt.ylabel('Number_of_Students')
plt.title('Students Exam score')
plt.show()
```

```python
#scatter plot
months=['Jan','Feb','Mar','April','May','June','July','Aug','Sept','Oct','Nov','Dec']
sales=[350,370,300,280,310,530,500,290,460,600,230,700]
plt.scatter(sales,months,color='black',marker='*',s=100)
plt.xlabel('Months')
plt.ylabel('Sales')
plt.title('Sales Vs Months Trend')
plt.show()
```

```python
#pie chart
#data
categories=['Electronics','Clothing','Grocery','Furniture','Books']
sales=[4500,3000,1500,2000,1000]
plt.pie(sales, labels=categories, shadow=True, autopct='%1.2f%%', explode=(0.1,0,0,0,0))
plt.title('Sales Distribution by Category')
plt.show()
```

```python
# stack plot
days=[1,2,3,4,5]

#data for activities
sleeping=[4,2,1,6,2]
eating=[3,4,6,7,1]
working=[2,4,6,1,3]

#plot
plt.stackplot(days,sleeping,eating,working,labels=['Sleeping','Eating','Working'])
plt.legend()
plt.show()
```

```python
#subplots
days=[1,2,3,4,5]

#data for activities
sleeping=[4,2,1,6,2]
eating=[3,4,6,7,1]
working=[2,4,6,1,3]

#create subplot1
plt.subplot(3,1,1)
plt.plot(days,sleeping)
plt.title("Sleeping")

#create subplot2
plt.subplot(3,1,2)
plt.plot(days,eating)
plt.title("Eating")

#create subplot3
plt.subplot(3,1,3)
plt.plot(days,working)
plt.title("Working")

plt.tight_layout()  # prevents overlapping
plt.show()
```
